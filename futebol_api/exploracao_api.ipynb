```python
def conteudo_api(link="/matches"): # Função que integra a aplicação na API e retorna o conteúdo
    url = "https://api.football-data.org/v4"
    link_api = f"{url}{link}"

    # Faz a solicitação do conteúdo da API
    headers = {'X-Auth-Token': token}
    obj_api = requests.get(link_api, headers=headers)
    # print(obj_api.status)

    return obj_api.json()
```
