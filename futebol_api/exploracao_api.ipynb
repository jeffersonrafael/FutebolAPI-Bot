```python
def conteudo_api(link="/matches"): # Fun√ß√£o que integra a aplica√ß√£o na API e retorna o conte√∫do
    url = "https://api.football-data.org/v4"
    link_api = f"{url}{link}"

    # Faz a solicita√ß√£o do conte√∫do da API
    headers = {'X-Auth-Token': token}
    obj_api = requests.get(link_api, headers=headers)
    # print(obj_api.status)

    return obj_api.json()
```

```python
def processamento_dados(dados): # Processa os dados recebidos da API
    # Usar regex pra converter as horas utc para utc-3
    formato_hora = r"\d{2}:\d{2}:\d{2}"
    formato_dia = r"\d{4}-\d{2}-\d{2}"

    for jogos in dados['matches']:
        
        horas = re.findall(formato_hora, jogos['utcDate'])
        dias = re.findall(formato_dia, jogos['utcDate'])


        if jogos['score']['winner'] == 'HOME_TEAM':
            resultado = f"\
üèÜ{jogos['homeTeam']['name']}: {jogos['score']['fullTime']['home']}\n\
    {jogos['awayTeam']['name']}: {jogos['score']['fullTime']['away']}\n\n\
"
        elif jogos['score']['winner'] == 'AWAY_TEAM':
            resultado = f"\
{jogos['homeTeam']['name']}: {jogos['score']['fullTime']['home']}\n\
    üèÜ{jogos['awayTeam']['name']}: {jogos['score']['fullTime']['away']}\n\n\
"
        else:
            resultado = f"\
{jogos['homeTeam']['name']}: {jogos['score']['fullTime']['home']}\n\
    {jogos['awayTeam']['name']}: {jogos['score']['fullTime']['away']}\n\n\
"


        if dias[0] in jogos['utcDate']:

            try:
                if jogos['referees'] != 0: 
                    arbitro = jogos['referees'][0]['name']
            except IndexError:
                arbitro = " N√£o informado "
        


            msg_info = f"\
    üìÖ Data da partida: {dias[0]}\n\
    üïî Hor√°rio da partida: {int(horas[0][0:2])-3}{horas[0][2:]}\n\
    ‚öΩ Competi√ß√£o: {jogos['competition']['name']}\n\
    ‚è≥ Situa√ß√£o da partida: {jogos['status']}\n\n\
    {resultado}\
    Arbitro: {arbitro}\n\
    =========================\n\n"


            print(msg_info)
        
        else:
            print(msg_info)
```
